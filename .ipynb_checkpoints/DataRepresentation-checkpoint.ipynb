```python
import matplotlib.pyplot as plt
import numpy as np
import csv

TimeData = []
TemperatureData = []
HumidityData = []
LPGData = []
MethaneData = []
CarbonMonoxideData = []

with open('DATALOG.csv', newline='') as csvfile:
    datareader = csv.reader(csvfile)
    for row in datareader:
        TimeData.append(row[0])
        TemperatureData.append(row[1])
        HumidityData.append(row[2])
        LPGData.append(row[3])
        MethaneData.append(row[4])
        CarbonMonoxideData.append(row[5])

TemperatureData    = list(map(float, TemperatureData))
HumidityData       = list(map(float, HumidityData))
LPGData            = list(map(float, LPGData))
MethaneData        = list(map(float, MethaneData))
CarbonMonoxideData = list(map(float, CarbonMonoxideData))

x = [i for i in range(len(TemperatureData))]

plt.subplot(211)
plt.plot(x, TemperatureData, x, HumidityData)

plt.subplot(212)
plt.plot(x, LPGData, x, MethaneData, x, CarbonMonoxideData)
```

 Здесь на первом графике оранжевым отмечена влажность воздуха, а синем его температура.
 На втором графике оранжевый это сжиженный углеводородный газ, синий метан, зеленый угарный газ.
 На графиках продемонстрировано изменение данных показателей в процессе проветривания комнаты.
